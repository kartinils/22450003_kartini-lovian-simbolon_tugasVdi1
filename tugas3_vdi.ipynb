```python
import numpy as np
import pandas as pd

df1 = pd.read_csv('diabetes_null.csv', na_values = ['#NAME'])
data = pd.DataFrame(df1)
data.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set(style="whitegrid")

# Set the figure size for better visualization
plt.figure(figsize=(10, 6))

# Create a boxplot to compare Glucose levels for patients with and without diabetes
plt.subplot(1, 2, 1)
boxplot_glucose = sns.boxplot(x='Outcome', y='Glucose', data=data, palette='coolwarm', hue='Outcome', legend=False)
boxplot_glucose.set_title('Glucose Levels by Diabetes Outcome', fontsize=14)
boxplot_glucose.set_xlabel('Diabetes Outcome (0 = No, 1 = Yes)', fontsize=12)
boxplot_glucose.set_ylabel('Glucose Level', fontsize=12)

# Create a boxplot to compare BMI for patients with and without diabetes
plt.subplot(1, 2, 2)
boxplot_bmi = sns.boxplot(x='Outcome', y='BMI', data=data, palette='coolwarm', hue='Outcome', legend=False)
boxplot_bmi.set_title('BMI by Diabetes Outcome', fontsize=14)
boxplot_bmi.set_xlabel('Diabetes Outcome (0 = No, 1 = Yes)', fontsize=12)
boxplot_bmi.set_ylabel('BMI', fontsize=12)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()

```

Tujuan:
Monitor Risiko Diabetes: Tujuan ini paling relevan, karena data tersebut berisi berbagai variabel yang mempengaruhi risiko diabetes, seperti glukosa, tekanan darah, dan BMI. Memantau faktor-faktor ini secara berkala membantu mencegah atau mengelola diabetes.
